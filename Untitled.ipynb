OCR Nedir?


OCR (Optical Character Recognition / Optik Karakter Tanıma)
OCR, bir görüntü (fotoğraf, taranmış belge, ekran görüntüsü vs.) içindeki yazılı metni tanıyıp, bunu makine tarafından okunabilir dijital metne dönüştüren teknolojidir.

PDF veya resim dosyasında yazı varsa, OCR ile bu yazıları metin dosyasına çevirebilirsin

En çok kullanılan örnek: Telefonunda bir yazıyı fotoğrafladığında, metni kopyalamana izin veren sistemler.

```python
boxes = results[0].boxes.xyxy.cpu().numpy()   # tespit edilen kutular
ocr_results = []

for box in boxes:
    x1, y1, x2, y2 = map(int, box)
    crop = img[y1:y2, x1:x2]                  # Kutuyu kırp
    if crop.size == 0:
        continue                             # Hatalı kutu atla
    # OCR uygula
    text_result = reader.readtext(crop)
    text = ' '.join([item[1] for item in text_result]) if text_result else ''
    ocr_results.append({'box': [x1, y1, x2, y2], 'text': text})

    # Kutunun içini göster
    plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
    plt.title(f'OCR: {text}')
    plt.axis('off')
    plt.show()
```
